```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
file_path = os.path.abspath('sale_sfr.csv')
med_sale_price_sfr = pd.read_csv(file_path, skiprows=1)
med_sale_price_sfr
```

```python
# Rename unnamed column
data_cleaned = med_sale_price_sfr.rename(columns={'Unnamed: 0': 'Region'})
data_cleaned
```

```python
# Reshape to long
data_melted = pd.melt(data_cleaned, id_vars=['Region'], var_name='Month', value_name='Sale Price')

# Clean data
data_melted['Sale Price'] = data_melted['Sale Price'].replace({'\$': '', 'K': '', ',': ''}, regex=True).astype(float) * 1000
```

```python
data_melted['Year'] = data_melted['Month'].str.extract(r'(\d{2})$')
data_melted['Year'] = '20' + data_melted['Year']  # Add '20' to get full year
```

```python
data_melted['Year'] = pd.to_numeric(data_melted['Year'])
data_melted
```

```python
plt.figure(figsize=(10, 6))

for region in data_melted['Region'].unique():
    region_data = data_melted[data_melted['Region'] == region]
    plt.plot(region_data['Month'], region_data['Sale Price'], label=region)

ax = plt.gca()
xticks = ax.get_xticks()
ax.set_xticks(xticks[::6])

plt.title('Median Sale Price for Single Family Zoned Homes in Bay Area Counties')
plt.xlabel('Month')
plt.ylabel('Median Sale Price ($)')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```
